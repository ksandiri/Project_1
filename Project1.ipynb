```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/Users/krishikasandiri/Downloads/Students Social Media Addiction.csv")
```

```python
print("First 5 rows of dataset:")
print(df.head())
```

```python
print("\nDataset Shape:", df.shape)
```

```python
print("\nColumns in dataset:")
print(df.columns)
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing Values Per Column:")
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")

print("\nCleaned Column Names:")
print(df.columns)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="avg_daily_usage_hours",
    y="conflicts_over_social_media"
)

plt.title("Daily Social Media Usage vs Conflicts Over Social Media")
plt.xlabel("Average Daily Usage (Hours)")
plt.ylabel("Conflicts Over Social Media")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="addicted_score",
    y="relationship_status"
)

plt.title("Addiction Level vs Relationship Satisfaction")
plt.xlabel("Self-Reported Addiction Level")
plt.ylabel("Relationship Satisfaction Score")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.regplot(
    data=df,
    x="addicted_score",
    y="mental_health_score",
)

plt.title("Addiction Score vs Mental Health Score")
plt.xlabel("Addiction Score")
plt.ylabel("Mental Health Score")

plt.show()
```
